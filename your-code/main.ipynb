#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4

* Explore plt.xticks

```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here
steps_by_month = fitbit.groupby('Months_encoded')['Steps'].median()
# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here
steps_workdays = fitbit.groupby('Work_or_Weekend')['Steps'].median()

fig, (ax1, ax2) = plt.subplots(1,2, figsize = (12,4))
plt.sca(ax1)
plt.plot(steps_by_month)
plt.legend('Median steps',loc = 'upper left')
plt.title('Median steps by month',fontsize=18)
plt.xlabel('Month',fontsize = 16)
plt.ylabel('Median steps',fontsize = 16)
plt.yticks(rotation=45)
plt.grid(True)

plt.sca(ax2)
plt.plot(steps_workdays,'o',color='red',markersize = 15)
plt.legend('Median steps',loc = 'upper left')
plt.title('Median steps by type of day',fontsize=18)
plt.xlabel('Workday (1 yes, 0 no)',fontsize = 16)
plt.ylabel('Median steps',fontsize = 16)
plt.yticks(rotation=45)
plt.grid(True)
plt.savefig('../Figures/Median steps.png')
plt.show()
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# your code here
f, ax = plt.subplots(1,3, figsize = (16,4))
x_values = fitbit['Steps'].values
y_values = [fitbit['Minutes Lightly Active'],fitbit['Minutes Very Active'],fitbit['Minutes Sedentary']]
titles = ['Minutes Lightly Active vs Steps','Minutes Very Active vs Steps','Minutes Sedentary vs Steps']
for i in range(3):
    plt.sca(ax[i])
    plt.plot(x_values,y_values[i],'o')
    plt.xlabel('Steps')
    plt.ylabel('Minutes')
    plt.title(titles[i])
plt.savefig('../Figures/Minutes vs Steps.png')
```

#### Plot the correlation matrix using seaborn.

```python
# your code here
# Compute the correlation matrix
corr = titanic.corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=np.bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=0.8, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5}).set_title("Correlation matrix")
plt.show()
print(corr)
```
